```python
#Import numpy,pandas packages.
import numpy as np
import pandas as pd
#Import seaborn and matplotlib for plots.
import seaborn as sns
import matplotlib.pyplot as plt
#Read the csv file and load the data into the dataframe titanic_df.
filename='titanic-data.csv' 
titanic_df=pd.read_csv(filename)
```

<p>Creating new Survival column indicating the survivability of passengers whether survived or not.</p>

```python
#Create survival label column
titanic_df['Survival']=titanic_df.Survived.map({0:'Died',1:'Survived'})
titanic_df.Survival.head()
```

<p>Creating new Class column containing the type of class used by passengers.</p>

```python
#Create Class label column
titanic_df['Class']=titanic_df.Pclass.map({1:'First Class',2:'Second Class',3:'Third Class'})
titanic_df.Class.head()
```

```python
titanic_df.isnull().head()#Finding any missing values
```

<p>It shows that there are missing values in column Age and Cabin.Here we are not considering the Cabin column in our investigation.But the age column is required.So we need to fill with some values for better result.</p>

```python
titanic_df.fillna(int(titanic_df['Age'].mean()),None,1)#Filling the column with missing values with the integer value of the mean of ages.
```

<p>We obtain the integer value of mean of ages is 29 which is filled in place of missing values.</p>

```python
titanic_df1=titanic_df.fillna(int(titanic_df['Age'].mean()),None,1)#Creating new dataframe with no missing values.
```

```python
age_df=titanic_df1['Age']
age_plot=age_df.plot(kind='hist',title='Age distribution')#plot showing the distribution of age of passengers.
age_plot.set_ylabel('Age')
age_plot.set_xlabel('Passenger')
```

<p1>Most of the passengers are in the age of 20-50.There are very few old age passengers.</p1>

```python
def diff_age_survival(age):
    """Function checking which age class of passengers are more."""
    if age<=20:
        return "child"
    elif age>20 and age<=60:
        return "middle age"
    elif age>60:
        return "old"
titanic_df1['Age_division']=titanic_df1['Age'].apply(diff_age_survival)
```

```python
titanic_df1['Age_division'].value_counts()
```

<p1>This information also shows that more passengers were there in middle age group.There are only few old age passengers.</p1>

```python
group_by_age=titanic_df1.groupby('Age_division').mean()# mean of age  of passengers who survived in various age groups.
group_by_age['Survived']
```

```python
import seaborn as sns#The plot shows the survivability based on Pclass and Sex.
ax = sns.violinplot(x="Age_division", y="Survived",data=titanic_df1)
ax.set_title('Survivability based on age')
```

<p1>The plot shows that child group were having more survival rate than other groups.</p1>

<h2>Standardise the Survived results</h2>

```python
def standardize_col(survive):
    res=(survive-survive.mean())/survive.std(ddof=0)#The survived records are standardized.
    '''
    To standardize a variable, convert each value to the
    number of standard deviations it is above or below the mean.
    '''
    return res
std=standardize_col(titanic_df1['Survived']).head()
print(std)
```
